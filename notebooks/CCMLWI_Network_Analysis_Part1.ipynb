1.8 Finding biggest hubs and plotting.

```python
hubs_mat = nx.algorithms.link_analysis.hits_alg.hub_matrix(g)
```

```python
len(hubs_mat)
```

```python
hubs_auth_vals = nx.algorithms.link_analysis.hits_alg.hits_numpy(g)
```

```python
hubs_sorted = sorted(hubs_auth_vals[0], key = lambda k: hubs_auth_vals[0][k])
```

```python
hubs_sorted
```

```python
g_hubs = nx.draw_networkx(g, pos, with_labels = True, node_color = 'yellow', edge_color = 'grey', labels = labels, node_size = [v * 50 for v in hubs_sorted], arrowsize = 0.2)
```
